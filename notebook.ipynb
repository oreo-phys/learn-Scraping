## 太宰治全集目次を取得する

・研究余録　～全集目次総覧～  
http://kenkyuyoroku.blog84.fc2.com/blog-entry-643.html

```python
from bs4 import BeautifulSoup
import urllib.request as req
```

GoogleChrome の開発ツールを用いて、CSSセレクタを取得する

<img width="1000" src="./img/homepage.png">

```python
url = "http://kenkyuyoroku.blog84.fc2.com/blog-entry-643.html"
# urlopen()でデータの取得
res = req.urlopen(url)
# BeautifulSoupでHTMLの解析
soup = BeautifulSoup(res, "html.parser")

li_list = soup.select_one("#center > div.ently_outline > div > div.ently_text")
print(li_list)
```

get_text()を使うことでタグに囲まれていない部分を取得することができる

```python
print(li_list.get_text())
```

作品名が「列車‥‥‥‥‥‥‥‥‥‥‥‥‥‥‥‥‥　3」のような表記になっているため、split()で分けて不要な部分を削除する

```python
data = li_list.get_text(",").split(",")

name_list = []
for text in data:
    name = text.split("‥")[0]
    name_list.append(name)

# 表示
for name in name_list:
    print(name)
```

## Excelに書き込む

```python
import openpyxl

path = "./dazai.xlsx"

wb = openpyxl.Workbook()
ws = wb.worksheets[0]

for i in range(len(name_list)-1):
    ws.cell(row=i+1, column=1).value = name_list[i]

wb.save(path)

```
